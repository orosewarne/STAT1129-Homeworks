```python
#question 4

x1 = np.array ([0,1,2,3,4])
y1 = np.array ([0,2,4,6,8])

x2 = np.array ([0,1,2,3,4])
y2 = np.array ([1,3,5,7,9])

x3 = np.array ([0,1,2,3,4])
y3 = np.array ([2, 4, 6, 8, 10])

plt.xlabel("x-label")
plt.ylabel("y-label")

plt.plot(x1,y1,x2,y2,x3,y3)
plt.show()
```

```python
#question 5 

x = np.array (['Andy', 'Amy', 'James', 'Jules', 'Arthur'])
y = np.array ([88, 66, 90, 55, 77])

plt.title("Student's Grades")
plt.bar(x,y)
plt.show()
```

```python
#question 5

y = np.array([88,66,90,55,77])
mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]

plt.title("Student's Grades")
plt.pie(y, labels = mylabels)
plt.legend(title = "Student's Names", bbox_to_anchor = (1,1), loc = 'upper left')
plt.show()
```

```python
#question 6

x1 = np.array ([0,1,2,3,4])
y1 = np.array ([0,2,4,6,8])

x2 = np.array ([0,1,2,3,4])
y2 = np.array ([1,3,5,7,9])

x3 = np.array ([0,1,2,3,4])
y3 = np.array ([2, 4, 6, 8, 10])

plt.xlabel("x-label")
plt.ylabel("y-label")

plt.subplot(2,3,1)
plt.plot(x1,y1,x2,y2,x3,y3)
plt.title("Plot 1")


x = np.array (['Andy', 'Amy', 'James', 'Jules', 'Arthur'])
y = np.array ([88, 66, 90, 55, 77])

plt.subplot(2,3,2)
plt.bar(x,y)
plt.title("Plot 2")

y = np.array([88,66,90,55,77])
mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]

plt.subplot(2,3,3)
plt.title("Plot 3")
plt.pie(y, labels = mylabels)


x = np.array([0,1,2,3,4])
y = np.array([2,4,6,8,10])

plt.subplot(2,3,4)
plt.plot(x,y)
plt.title("Plot 4")


x = np.array([1,2,3,4])
y = np.array([2,6,4,5])

plt.subplot(2,3,5)
plt.plot(x,y,'o')
plt.title("Plot 5")


x = [10,2,3,10,6,2,3,4,3,2]

plt.subplot(2,3,6)
plt.hist(x)
plt.title("Plot 6")

plt.show()
```
